Amazon's product reviews play a crucial role in driving customer purchase decisions and building trust in the e-commerce platform. However, the current system of relying on organic customer votes to determine a review's helpfulness can be slow and inefficient, especially for new products or reviews. This delay in identifying truly helpful reviews can lead to suboptimal customer experiences and hinder informed decision-making.

The key problems that Amazon aims to address with an automated review helpfulness prediction tool are:

### **Slow Discovery of Helpful Reviews**
New product reviews take time to accumulate enough votes to gauge their helpfulness objectively.
During this period, customers may miss out on highly insightful reviews that could aid their purchase decisions.
### **Inefficient Surfacing of Relevant Content**
With the vast volume of reviews, helpfulness signals are diluted, making it challenging for customers to find the most relevant and useful information quickly.
Potential buyers may need to sift through numerous reviews, including irrelevant or low-quality ones, to make informed decisions.
### **Suboptimal Customer Experience**
The delay in identifying helpful reviews can lead to frustration and dissatisfaction among customers seeking comprehensive product information.
Customers may make uninformed purchases or abandon potential transactions due to a lack of easily accessible, high-quality review content.
### **Missed Opportunities for Product Improvement**
Insightful reviews often contain valuable feedback and suggestions for product enhancements.
By not prioritizing these reviews effectively, Amazon may miss opportunities to identify areas for improvement and address customer concerns promptly.


By developing an automated tool that can accurately predict the long-term helpful vote count of a product review soon after it is posted, Amazon can address these problems and unlock several potential benefits:

- Enhance customer experience by surfacing the most helpful reviews prominently from the outset, enabling informed purchase decisions.
- Increase customer trust and confidence in the review system by ensuring high-quality, insightful content is readily available.
- Improve operational efficiency by automating the process of identifying and prioritizing helpful reviews, reducing manual efforts.
- Gain faster insights into product strengths, issues, and areas for improvement based on the prioritized helpful reviews.
- Maintain a competitive advantage by providing a superior review experience compared to other e-commerce platforms.

The success of this automated tool lies in its ability to accurately learn the patterns and characteristics that make a review truly helpful, leveraging machine learning or deep learning techniques on Amazon's vast repository of historical review data. By addressing the problem of slow helpful review discovery and inefficient content surfacing, Amazon can unlock significant value for both customers and the business.

Benefits and Challenges
This automated tool for predicting helpful review votes could provide several benefits for Amazon and its customers:

- Improved customer experience by surfacing the most helpful reviews early on, aiding purchase decisions.
- Increased trust and confidence in the review system by prioritizing high-quality, insightful reviews.
- Potential for better product feedback and improvements based on the identified characteristics of helpful reviews.

However, some challenges might include:

- Ensuring model fairness and absence of bias, as review helpfulness can be subjective and influenced by various factors.
- Maintaining model accuracy over time as language patterns, product categories, and customer preferences evolve.
- Balancing the prioritization of predicted helpful reviews with the organic discovery of new, potentially helpful reviews.
- Addressing privacy and ethical considerations regarding the use of customer data for training and inference.

##*Outliers & Skewness*


```python
# Code for skewness correction (see source below)
# Depending upon the characteritics of a feature (column), a log, Box-Cox or power transform is applied to normalize the distribution

# -*- coding: utf-8 -*-
"""
Created on Sat Feb 23 14:42:46 2019
@author: DATAmadness
"""

##################################################
# A function that will accept a pandas dataframe
# and auto-transforms columns that exceeds threshold value
#  -  Offers choice between boxcox or log / exponential transformation
#  -  Automatically handles negative values
#  -  Auto recognizes positive /negative skewness

# Further documentation available here:
# https://datamadness.github.io/Skewness_Auto_Transform

def skew_autotransform(DF, include = None, exclude = None, plot = False, threshold = 1, exp = False):

    #Get list of column names that should be processed based on input parameters
    if include is None and exclude is None:
        colnames = DF.columns.values
    elif include is not None:
        colnames = include
    elif exclude is not None:
        colnames = [item for item in list(DF.columns.values) if item not in exclude]
    else:
        print('No columns to process!')

    #Helper function that checks if all values are positive
    def make_positive(series):
        minimum = np.amin(series)
        #If minimum is negative, offset all values by a constant to move all values to positive teritory
        if minimum <= 0:
            series = series + abs(minimum) + 0.01
        return series


    #Go through desired columns in DataFrame
    for col in colnames:
        #Get column skewness
        skew = DF[col].skew()
        transformed = True

        if plot:
            #Prep the plot of original data
            sns.set_style("darkgrid")
            sns.set_palette("Blues_r")
            fig, axes = plt.subplots(1, 2, figsize=(10, 5))
            #ax1 = sns.distplot(DF[col], ax=axes[0])
            ax1 = sns.histplot(DF[col], ax=axes[0], color="blue", label="100% Equities", kde=True, stat="density", linewidth=0)
            ax1.set(xlabel='Original ' + str(col))

        #If skewness is larger than threshold and positively skewed; If yes, apply appropriate transformation
        if abs(skew) > threshold and skew > 0:
            skewType = 'positive'
            #Make sure all values are positive
            DF[col] = make_positive(DF[col])

            if exp:
               #Apply log transformation
               DF[col] = DF[col].apply(math.log)
            else:
                #Apply boxcox transformation
                DF[col] = ss.boxcox(DF[col])[0]
            skew_new = DF[col].skew()

        elif abs(skew) > threshold and skew < 0:
            skewType = 'negative'
            #Make sure all values are positive
            DF[col] = make_positive(DF[col])

            if exp:
               #Apply exp transformation
               DF[col] = DF[col].pow(10)
            else:
                #Apply boxcox transformation
                DF[col] = ss.boxcox(DF[col])[0]
            skew_new = DF[col].skew()

        else:
            #Flag if no transformation was performed
            transformed = False
            skew_new = skew

        #Compare before and after if plot is True
        if plot:
            print('\n ------------------------------------------------------')
            if transformed:
                print('\n %r had %r skewness of %2.2f' %(col, skewType, skew))
                print('\n Transformation yielded skewness of %2.2f' %(skew_new))
                sns.set_palette("Paired")
                #ax2 = sns.distplot(DF[col], ax=axes[1], color = 'r')
                ax2 = sns.histplot(DF[col], ax=axes[1], color="red", label="100% Equities", kde=True, stat="density", linewidth=0)
                ax2.set(xlabel='Transformed ' + str(col))
                plt.show()
            else:
                print('\n NO TRANSFORMATION APPLIED FOR %r . Skewness = %2.2f' %(col, skew))
                #ax2 = sns.distplot(DF[col], ax=axes[1])
                ax2 = sns.histplot(DF[col], ax=axes[1], color="blue", label="100% Equities", kde=True, stat="density", linewidth=0)
                ax2.set(xlabel='NO TRANSFORM ' + str(col))
                plt.show()


    return DF
```

```python
# Use code above (adapted from https://github.com/datamadness/Automatic-skewness-transformation-for-Pandas-DataFrame) to correct skewness
# All the predictors are real-valued, so we can push them all through the skewness check/correction.
X_quant = skew_autotransform(X_quant.copy(deep=True), plot = True, exp = False, threshold = 1)
```

```python
# Tukey Rule outliers
# As an alternative, you could use z-scores greater than 3 or less than -3.

cols = X_quant.columns
#Tukey's method
def tukey_rule(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    upper_lim = data[col].quantile(0.5) + 2 * IQR
    lower_lim = data[col].quantile(0.5) - 2 * IQR
    outliers = []
    for index, x in enumerate(data[col]):
        if x < lower_lim or x >= upper_lim:
            outliers.append(index)
    return outliers

# Identify outliers
for i in cols:
  outliers_Tukey = tukey_rule(X_quant,i)
  print("Column ",i,": ",outliers_Tukey)

# Windsorize X and check the results
print("Descriptive Statistics Before", X_quant.describe())
X_quant_winsorized = X_quant.copy(deep=True)
for i in cols:
  X_quant_winsorized[i] = winsorize(X_quant[i], limits=(0.05, 0.05))
print("Descriptive Statistics After", X_quant_winsorized.describe())
```

```python
# Regressor Development and Assessment

names = ['Multiple Linear Regression', 'Lasso Regression', 'Ridge Regression','ElasticNet Regression','kNN', 'SVM','Decision Tree', 'Multilayer Perceptron','Random Forest', 'XGBoost', 'AdaBoost','LightGBM']
regressors = [linear_model.LinearRegression(),
              Lasso(alpha=0.5),
              Ridge(alpha=0.5),
              ElasticNet(alpha=0.25, l1_ratio=0.5),
              KNeighborsRegressor(n_neighbors=5),
              SVR(kernel = 'poly'),   #Can try kernels 'poly', 'linear', 'rbf' and 'sigmoid' kernels
              DecisionTreeRegressor(),
              MLPRegressor(hidden_layer_sizes=(25,25,1),
                       max_iter = 1000,activation = 'relu',
                       solver = 'adam'),
              RandomForestRegressor(criterion='squared_error'),
              XGBRegressor(objective ='reg:squarederror'),
              AdaBoostRegressor(loss='square'),
              lgb.LGBMRegressor(objective='mean_squared_error')
              ]

for name, regr in zip(names, regressors):
  print('REGRESSOR: ',name)
  pipe = Pipeline(steps=[('normalize', MinMaxScaler()), ('model', regr)])
  model = TransformedTargetRegressor(regressor=pipe, transformer=MinMaxScaler())
  cv = KFold(n_splits=10, shuffle=True, random_state=1)
  scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=cv, n_jobs=-1)
  score_mean = mean(absolute(scores))
  score_stdev = np.std(scores)
  print('Mean (std Dev) of MSE: %.3f (%.3f)\n' % (score_mean, score_stdev))
```

# Delta of base learners vs ensembles

```python
from sklearn.ensemble import AdaBoostRegressor
import lightgbm as lgb

# Base Learners
MLR = linear_model.LinearRegression()
lasso = Lasso(alpha=0.5)
ridge = Ridge(alpha=0.5)
elasticnet = ElasticNet(alpha=0.25, l1_ratio=0.5)

# Ensemble Methods
ada_boost = AdaBoostRegressor(base_estimator=DecisionTreeRegressor(), loss='square')
lightgbm = lgb.LGBMRegressor(objective='mean_squared_error')

# Train base learners and ensemble methods
base_learners = [MLR, lasso, ridge, elasticnet]
ensemble_methods = [ada_boost, lightgbm]

base_learner_scores = []
ensemble_method_scores = []

for learner in base_learners:
    pipe = Pipeline(steps=[('normalize', MinMaxScaler()), ('model', learner)])
    model = TransformedTargetRegressor(regressor=pipe, transformer=MinMaxScaler())
    cv = KFold(n_splits=10, shuffle=True, random_state=1)
    scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=cv, n_jobs=-1)
    base_learner_scores.append(scores.mean())

for ensemble in ensemble_methods:
    model = TransformedTargetRegressor(regressor=ensemble, transformer=MinMaxScaler())
    cv = KFold(n_splits=10, shuffle=True, random_state=1)
    scores = cross_val_score(model, X, y, scoring='neg_mean_squared_error', cv=cv, n_jobs=-1)
    ensemble_method_scores.append(scores.mean())

# Calculate delta
base_learner_avg_score = np.mean(base_learner_scores)
ensemble_method_avg_score = np.mean(ensemble_method_scores)

delta_base_vs_ensemble = ensemble_method_avg_score - base_learner_avg_score

print("Delta of base learner vs ensemble:", delta_base_vs_ensemble)
```

# Relative performance of boosting vs DL vs AutoML

```python
from sklearn.neural_network import MLPRegressor
from xgboost import XGBRegressor

# Boosting (AdaBoost) vs Deep Learning (MLP) vs AutoML (XGBoost)

# Train boosting (AdaBoost)
ada_boost = AdaBoostRegressor(loss='square')
cv = KFold(n_splits=10, shuffle=True, random_state=1)
scores_ada = cross_val_score(ada_boost, X, y, scoring='neg_mean_squared_error', cv=cv, n_jobs=-1)

# Train deep learning (MLP)
mlp = MLPRegressor(hidden_layer_sizes=(25, 25, 1), max_iter=1000, activation='relu', solver='adam')
cv = KFold(n_splits=10, shuffle=True, random_state=1)
scores_mlp = cross_val_score(mlp, X, y, scoring='neg_mean_squared_error', cv=cv, n_jobs=-1)

# Train AutoML (XGBoost)
xgboost = XGBRegressor(objective='reg:squarederror')
cv = KFold(n_splits=10, shuffle=True, random_state=1)
scores_xgboost = cross_val_score(xgboost, X, y, scoring='neg_mean_squared_error', cv=cv, n_jobs=-1)

# Calculate mean absolute average scores
avg_score_ada = np.mean(scores_ada)
avg_score_mlp = np.mean(scores_mlp)
avg_score_xgboost = np.mean(scores_xgboost)

print("Relative performance of boosting vs DL vs AutoML:")
print("AdaBoost Average Score:", avg_score_ada)
print("MLP Average Score:", avg_score_mlp)
print("XGBoost Average Score:", avg_score_xgboost)
```
